## Preprocessing

```python
# Clear filenames (strip path and extension)
def clear_filename(filename):
    start_offset = len("c:\\Users\\Joe\\Test\\o\\Day1\\")
    end_offset = len(".jpg")
    return filename[start_offset:][:-end_offset]

for dataset in [ob1images, ob2images, ob3images, ob4images, ob5images, ob6images,
                ob11images, ob12images, ob13images, ob14images, ob15images,
                ob16images, ob17images, ob18images, ob19images]:
    dataset.Image = dataset.Image.apply(func=clear_filename)    
```

```python
# Translate all timestamps to milliseconds
import dateutil
import datetime

def convert_psychopy_timestamp(timestamp_string):
    dt = dateutil.parser.parse(timestamp_string)  # curr time
    epoch = datetime.datetime.utcfromtimestamp(0) # timestamp 0
    ts = (dt - epoch).total_seconds() * 1000.0    # get time difference in milliseconds
    return ts

# PsychoPy timestamps ('2017-04-26 13:26:19.479000')
for dataset in [ob1images, ob2images, ob3images, ob4images, ob5images, ob6images,
                ob11images, ob12images, ob13images, ob14images, ob15images,
                ob16images, ob17images, ob18images, ob19images]:
    dataset.Timestamp = dataset.Timestamp.apply(func=convert_psychopy_timestamp)

# Empatica timestamps (in seconds: 1493212602.4958858); add 2 hours for timezone difference
for dataset in [ob4hr, ob4gsr, ob5hr, ob5gsr, ob13hr, ob13gsr, ob18hr, ob18gsr]:
    dataset.Timestamp = dataset.Timestamp.apply(func=lambda t: 1000.0 * (t + 2*60*60))

# MsBand timestamps (in milliseconds: 1493205952723); add 2 hours for timezone difference
for dataset in [ob1hr, ob1gsr, ob2hr, ob2gsr, ob3hr, ob3gsr, ob6hr, ob6gsr,
                ob11hr, ob11gsr, ob12hr, ob12gsr, ob14hr, ob14gsr, ob15hr, ob15gsr,
                ob16hr, ob16gsr, ob17hr, ob17gsr, ob19hr, ob19gsr]:
    dataset.Timestamp = dataset.Timestamp.apply(func=lambda t: t + 2*60*60*1000)
```

```python
# Empatica: convert from microSiemes to kOhms == convert to siemens (val/1000000),
#           then convert to Ohms (1/val), and finally convert to kOhms (val/100)
ob4gsr.Value = 1/(ob4gsr.Value/1000000)/100
ob5gsr.Value = 1/(ob5gsr.Value/1000000)/100
ob13gsr.Value = 1/(ob13gsr.Value/1000000)/100
ob18gsr.Value = 1/(ob18gsr.Value/1000000)/100
```

```python
# for each subject*sensor create a list (with relative timestamps):
# [
#     (image, [timestamp1, timestamp2, ...], [value1, value2, ...] ),
#     ...
# ]

ob1img_hr = []
ob1img_gsr = []
ob2img_hr = []
ob2img_gsr = []
ob3img_hr = []
ob3img_gsr = []
ob4img_hr = []
ob4img_gsr = []
ob5img_hr = []
ob5img_gsr = []
ob6img_hr = []
ob6img_gsr = []
ob11img_hr = []
ob11img_gsr = []
ob12img_hr = []
ob12img_gsr = []
ob13img_hr = []
ob13img_gsr = []
ob14img_hr = []
ob14img_gsr = []
ob15img_hr = []
ob15img_gsr = []
ob16img_hr = []
ob16img_gsr = []
ob17img_hr = []
ob17img_gsr = []
ob18img_hr = []
ob18img_gsr = []
ob19img_hr = []
ob19img_gsr = []

# time difference (in ms) between marker and the end of the recording for image (8 seconds)
tdiff = 8 * 1000.0

for dataset in [(ob1images, ob1hr, ob1img_hr), (ob1images, ob1gsr, ob1img_gsr),
               (ob2images, ob2hr, ob2img_hr), (ob2images, ob2gsr, ob2img_gsr),
               (ob3images, ob3hr, ob3img_hr), (ob3images, ob3gsr, ob3img_gsr), 
               (ob4images, ob4hr, ob4img_hr), (ob4images, ob4gsr, ob4img_gsr), 
               (ob5images, ob5hr, ob5img_hr), (ob5images, ob5gsr, ob5img_gsr), 
               (ob6images, ob6hr, ob6img_hr), (ob6images, ob6gsr, ob6img_gsr),
               (ob11images, ob11hr, ob11img_hr), (ob11images, ob11gsr, ob11img_gsr),
               (ob12images, ob12hr, ob12img_hr), (ob12images, ob12gsr, ob12img_gsr),
               (ob13images, ob13hr, ob13img_hr), (ob13images, ob13gsr, ob13img_gsr),
               (ob14images, ob14hr, ob14img_hr), (ob14images, ob14gsr, ob14img_gsr),
               (ob15images, ob15hr, ob15img_hr), (ob15images, ob15gsr, ob15img_gsr),
               (ob16images, ob16hr, ob16img_hr), (ob16images, ob16gsr, ob16img_gsr),
               (ob17images, ob17hr, ob17img_hr), (ob17images, ob17gsr, ob17img_gsr),
               (ob18images, ob18hr, ob18img_hr), (ob18images, ob18gsr, ob18img_gsr),
               (ob19images, ob19hr, ob19img_hr), (ob19images, ob19gsr, ob19img_gsr)]:
    for img in dataset[0].itertuples():
        selected_rows = dataset[1].loc[(dataset[1]['Timestamp'] >= img.Timestamp) & (dataset[1]['Timestamp'] <= img.Timestamp + tdiff)]
        selected_rows.Timestamp = selected_rows.Timestamp.apply(func=lambda t: t - img.Timestamp)
        ts_list = []
        val_list = []
        for reading in selected_rows.itertuples():
            ts_list.append(reading.Timestamp)
            val_list.append(reading.Value)
        dataset[2].append( (img.Image, ts_list, val_list) )
```

```python
# prepare dataframes for valence and arousal groups:
# header: [ imageX1, imageX2, imageX3, ... ]
# index: timestamp

neutral_hr = pandas.DataFrame()
neutral_gsr = pandas.DataFrame()
valH_aroH_hr = pandas.DataFrame()
valH_aroH_gsr = pandas.DataFrame()
valH_aroL_hr = pandas.DataFrame()
valH_aroL_gsr = pandas.DataFrame()
valL_aroH_hr = pandas.DataFrame()
valL_aroH_gsr = pandas.DataFrame()

# HR data:
for dataset in [ob1img_hr, ob2img_hr, ob3img_hr, ob4img_hr, ob5img_hr, ob6img_hr,
               ob11img_hr, ob12img_hr, ob13img_hr, ob14img_hr, ob15img_hr, 
               ob16img_hr, ob17img_hr, ob18img_hr, ob19img_hr]:
    for image in dataset:
        # prepare DataFrame with Timestamps and HR Values for specific image within specific subject
        new_df = pandas.DataFrame(data={"Timestamp": pandas.to_timedelta(image[1], unit='ms'), image[0]: image[2]})
        new_df = new_df.set_index("Timestamp")
        
        # get image info (especially Valence and Arousal)
        curr_image = images_params.loc[images_params.ID == image[0]].squeeze()
        
        try:
            # concatenate with specific group
            if curr_image.Valence > 5.0 and curr_image.Arousal > 5.0:
                valH_aroH_hr = pandas.concat([valH_aroH_hr, new_df], axis=1)
            elif curr_image.Valence > 5.0 and curr_image.Arousal < 3.0:
                valH_aroL_hr = pandas.concat([valH_aroL_hr, new_df], axis=1)
            elif curr_image.Valence < 3.0 and curr_image.Arousal > 5.0:
                valL_aroH_hr = pandas.concat([valL_aroH_hr, new_df], axis=1)
            else:
                neutral_hr = pandas.concat([neutral_hr, new_df], axis=1)
        except:
            # FIXME: pojawiaja sie jakies bledy przy 54 probkach -- pewnie problem z jedna z osob badanych,
            # ale nie udalo mi sie zdiagnozowac tego problemu...
            print("Error")
            
# same for GSR data:
for dataset in [ob1img_gsr, ob2img_gsr, ob3img_gsr, ob4img_gsr, ob5img_gsr, ob6img_gsr,
               ob11img_gsr, ob12img_gsr, ob13img_gsr, ob14img_gsr, ob15img_gsr, 
               ob16img_gsr, ob17img_gsr, ob18img_gsr, ob19img_gsr]:
    for image in dataset:
        # prepare DataFrame with Timestamps and HR Values for specific image within specific subject
        new_df = pandas.DataFrame(data={"Timestamp": pandas.to_timedelta(image[1], unit='ms'), image[0]: image[2]})
        new_df = new_df.set_index("Timestamp")
        
        # get image info (especially Valence and Arousal)
        curr_image = images_params.loc[images_params.ID == image[0]].squeeze()
        
        try:
            # concatenate with specific group
            if curr_image.Valence > 5.0 and curr_image.Arousal > 5.0:
                valH_aroH_gsr = pandas.concat([valH_aroH_gsr, new_df], axis=1)
            elif curr_image.Valence > 5.0 and curr_image.Arousal < 3.0:
                valH_aroL_gsr = pandas.concat([valH_aroL_gsr, new_df], axis=1)
            elif curr_image.Valence < 3.0 and curr_image.Arousal > 5.0:
                valL_aroH_gsr = pandas.concat([valL_aroH_gsr, new_df], axis=1)
            else:
                neutral_gsr = pandas.concat([neutral_gsr, new_df], axis=1)
        except:
            print("Error")
```

```python
# generate the mean time-series for each group

def fill_missing_values(data):
    for col in range(data.shape[1]):
        col_vals = data.ix[:,col]
        last_val = 0
        nan_count = 0
        for val in range(len(col_vals)):
            if pandas.isnull(col_vals[val]) or (abs(col_vals[val]) < 0.0000001):
                nan_count += 1
                continue
            if nan_count > 0:
                diff_val = (col_vals[val] - last_val) / (nan_count+1)
                for idx in range(1,nan_count+1):
                    col_vals[val-idx] = col_vals[val] - idx*diff_val
                nan_count = 0
            last_val = col_vals[val]
        data.ix[:,col] = col_vals
    return data


def calculate_mean(dataset):
    # resample the data that for each second ('s') there is only one probe (mean of the probes)
    result = dataset.resample('s').mean()
    
    # if there are seconds without any probe, fill it with the mean value of neighbours
    result = fill_missing_values(result)
    
    # calculate the mean value time-series across all stimulus within frame
    return result.mean(axis=1)


neutral_hr_MEAN = calculate_mean(neutral_hr)
neutral_gsr_MEAN = calculate_mean(neutral_gsr)
valH_aroH_hr_MEAN = calculate_mean(valH_aroH_hr)
valH_aroH_gsr_MEAN = calculate_mean(valH_aroH_gsr)
valH_aroL_hr_MEAN = calculate_mean(valH_aroL_hr)
valH_aroL_gsr_MEAN = calculate_mean(valH_aroL_gsr)
valL_aroH_hr_MEAN = calculate_mean(valL_aroH_hr)
valL_aroH_gsr_MEAN = calculate_mean(valL_aroH_gsr)
```
